```python
# Ejemplo conceptual de descenso de gradiente
x = 10
learning_rate = 0.1

for _ in range(10):
    grad = 2 * x  # derivada de x^2
    x -= learning_rate * grad
    print(f"x: {x}")
```

```python
import pandas as pd

# Crear datos ficticios
data = {
    'edad': [22, 25, 30, 35, 40, 28, 33, 45, 50, 27],
    'salario': [28000, 32000, 40000, 45000, 50000, 35000, 42000, 55000, 60000, 33000]
}

df = pd.DataFrame(data)

# Estadísticas básicas
media_edad = df['edad'].mean()
mediana_edad = df['edad'].median()
std_edad = df['edad'].std()

media_salario = df['salario'].mean()
mediana_salario = df['salario'].median()
std_salario = df['salario'].std()

# Mostrar resultados
print("Media edad:", media_edad)
print("Mediana edad:", mediana_edad)
print("Desviación estándar edad:", std_edad)

print("\nMedia salario:", media_salario)
print("Mediana salario:", mediana_salario)
print("Desviación estándar salario:", std_salario)

# Correlación entre edad y salario
correlacion = df['edad'].corr(df['salario'])
print("\nCorrelación edad - salario:", correlacion)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simular ingresos (media = 40,000, desviación = 8,000)
ingresos = np.random.normal(loc=40000, scale=8000, size=1000)

# Histograma
plt.figure(figsize=(10, 6))
plt.hist(ingresos, bins=30, color='skyblue', edgecolor='black')
plt.title("Distribución de Ingresos Mensuales", fontsize=14)
plt.xlabel("Ingreso ($)", fontsize=12)
plt.ylabel("Frecuencia", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Estadísticas extra (opcional)
print("Ingreso promedio:", np.mean(ingresos))
print("Desviación estándar de ingresos:", np.std(ingresos))
```
